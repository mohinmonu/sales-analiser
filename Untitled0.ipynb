```python
# 📦 Install libraries (first time only)
!pip install matplotlib pandas

# 🧠 Import tools
import pandas as pd
import matplotlib.pyplot as plt

# 📁 Create sample data (you can upload your own CSV later)
data = {
    'Date': ['2024-06-01', '2024-06-01', '2024-06-01'],
    'Item': ['Pepsi', 'Lays', 'Biscuits'],
    'Qty': [3, 5, 2],
    'Price': [30, 50, 20]
}

df = pd.DataFrame(data)

# 📊 Total sales per item
df['Total'] = df['Qty'] * df['Price']
sales_by_item = df.groupby('Item')['Total'].sum()

# 🔥 Plot bar chart
sales_by_item.plot(kind='bar', title='Sales by Item (₹)')
plt.xlabel("Item")
plt.ylabel("Total Sales ₹")
plt.show()
```
